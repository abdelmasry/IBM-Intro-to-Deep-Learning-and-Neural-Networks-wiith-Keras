## Concrete Data

<strong>The dataset is about the compressive strength of different samples of concrete based on the volumes of the different ingredients that were used to make them. Ingredients include:</strong>

<strong>1. Cement</strong>

<strong>2. Blast Furnace Slag</strong>

<strong>3. Fly Ash</strong>

<strong>4. Water</strong>

```python
import pandas as pd

concrete_data = pd.read_csv('concrete_data.csv')
concrete_data.head()
```

```python
concrete_data_columns = concrete_data.columns

Features = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']]
target = concrete_data['Strength']
```

```python
Features.head()
```

```python
target.head()
```

```python
n_cols = Features.shape[1]
print(n_cols)
```

### 0-Neural Network Model-0

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
def base_model():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(1))
    # compile model
    model.compile(optimizer='Adam', loss='mean_squared_error')
    return model
```

###
<strong> Splitting Data into 70% Train and 30% Test


```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(Features, target, test_size=0.3, random_state=4)
```

# Building NN Model

```python
NeuralNet = base_model()
```

```python
NeuralNet.fit(X_train, y_train, epochs=50, verbose=1)
```

```python
loss = NeuralNet.evaluate(X_test, y_test, verbose=0)
y_pred = NeuralNet.predict(X_test)
print("Validation Loss: ", loss)
```

## Mean Squared Error

```python
from sklearn.metrics import mean_squared_error
import numpy as np

mean_square_error = mean_squared_error(y_test, y_pred)
mean = np.mean(mean_square_error)
standard_deviation = np.std(mean_square_error)
print(mean)
```

<strong> List of Errors

```python
mean_squared_errors = []

for i in range(0, 50):
    X_train, X_test, y_train, y_test = train_test_split(Features, target, test_size=0.3, random_state=i)
    NeuralNet.fit(X_train, y_train, epochs=50, verbose=1)
    MSE = NeuralNet.evaluate(X_test, y_test, verbose=0)
    print("MSE " + str(i + 1) + ": " + str(MSE))
    y_pred = NeuralNet.predict(X_test)
    mean_square_error = mean_squared_error(y_test, y_pred)
    mean_squared_errors.append(mean_square_error)

mean_squared_errors = np.array(mean_squared_errors)
mean = np.mean(mean_squared_errors)
standard_deviation = np.std(mean_squared_errors)
```

<strong># Report

```python
print("Mean: " + str(mean))
print("Standard Deviation: " + str(standard_deviation))
```
